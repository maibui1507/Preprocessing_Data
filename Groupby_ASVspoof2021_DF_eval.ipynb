```python
import pandas as pd

# Khởi tạo dữ liệu mẫu
data = pd.read_csv("/content/drive/MyDrive/asv_df_eval_2021.txt", sep=" ")


# Gom nhóm dữ liệu theo giá trị "nocodec" và "notrim"
groups = data.groupby(["nocodec", "spoof", "A14", "notrim", "traditional_vocoder"])

data.info()
```

```python
#labeling data
data['type'] = None

for i, group in enumerate(groups.groups):
    common_indices = groups.get_group(group).index.intersection(data.index)
    data.loc[common_indices, 'type'] = i

data.head(5)
```
